```python
#Answer the Pandas exercises
# Exercise 1.2
#Get the First_Party console that has no sales.
gameSalesMerge = pd.merge(gameSales,firstParty,how='outer',on=('Platform'), indicator=True)
gameSalesMerge = gameSalesMerge[gameSalesMerge["First_Party"].notna()]
gameSalesMerge = gameSalesMerge[gameSalesMerge["Global_Sales"].isna()]
display(gameSalesMerge)
```

```python
#Answer the Pandas exercises
# Exercise 2
#Cleans the dataset of numerical errors (NaN) by imputing the average value of each column to the missing data.
gameSalesMerge = pd.merge(gameSales,firstParty,how='outer',on=('Platform'), indicator=True)

NumericCols = ["NA_Sales", "EU_Sales", "JP_Sales", "Global_Sales", "Other_Sales"]

for column in NumericCols:
    media = gameSalesMerge[column].mean()
    gameSalesMerge[column] = gameSalesMerge[column].fillna(media)

gameSalesMerge = gameSalesMerge[gameSalesMerge.isna().any(axis=1)]

display(gameSalesMerge)
```
